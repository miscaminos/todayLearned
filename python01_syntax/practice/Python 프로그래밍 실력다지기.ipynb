```python
# (3) reversed(dictview) - returns a reverse iterator over keys, values, items 
# 주의!! reversed() only works for python 3.8 or later
d = {'g': 1, 'b': 5, 'c': 3, 'd': 4}
k=reversed(d.keys())
print (list(k))#Output:['d', 'c', 'b', 'g']

v=reversed(d.values())
print (list(v))#Output:[4, 3, 5, 1]

i=reversed(d.items())
print (list(i))#Output:[('d', 4), ('c', 3), ('b', 5), ('g', 1)]

#return a reversed iterator of keys alone
d1=reversed(d)
print (list(d1))#Output:['d', 'c', 'b', 'g']
```

```python
def writelog(logfile, n, log):
    # localtime(): time.struct_time형식으로 출력된다.
    time_stamp = strftime('%Y-%m-%d %X\t', localtime())
    log = time_stamp + str(n) + log + '\n'
    with open(logfile,'a',encoding='utf-8') as f:
        f.writelines(log)
    
writelog(logfile, n, '번째 로깅 문장입니다.')
n+=1
```

```python
# stack 구현하기
# 스택은 차곡차곡 쌓이듯이 올라가서 나중에 저장된 자료가 먼저 추출될 수 
#  있도록 되어 있는 나열구조이다. LIFO: Last-In, First-Out
mystack = []

def putdata(data):
    global mystack 
    mystack.append(data)
    
def popdata():
    global mystack
    if len(mystack) == 0:
        return None
    return mystack.pop()
```

```python
putdata('data1')
putdata([3,4,5,6,7])
putdata(12345)
```

```python
#문장에 나타나는 문자 빈도수 계산하기
# getTextFreq(filename): 파일이름을 인자로 받아 이 파일에 있는 내용에서 각 문자들이
# 나타내는 비도수를 문자별로 계산하여 결과를 리턴

def getTextFreq(filename):
    with open(filename, 'r') as f:
        text = f.read()
        text= text.lower()
        fdict = {}
        for c in text: #문자 하나씩 추출
            if c in fdict:
                fdict[c] += 1
            else:
                fdict[c] = 1
    return fdict
```

```python
#파일에서 특정 단어의 개수 계산하기
#참고: find()는 찾고자 하는 문자열이 없을 경우 -1를 리턴한다.
def countWord(filename, word):
    with open(filename,'r',encoding='utf-8') as f:
        text = f.read()
        pos = text.find(word)
        count = 0
        while pos != -1:
            count+=1
            pos = text.find(word, pos+1)
    return count
```

```python
#용량이 큰 파일을 1MB 파일 10개로 분리하기
filename = './files/python-3.5.2.exe'
subsize = 1024*1024*3 #3MB
suffix = 0 #분리될 파일 이름 뒤에 추가할 숫자

with open(filename, 'rb') as f:
    buf = f.read(subsize)
    while buf:
        subfilename = filename + '_'+str(suffix)
        with open(subfilename,'wb') as h:
            h.write(buf)
            print('[%s]완료' %subfilename)
        buf = f.read(subsize)
        suffix+=1
```

```python
#파일을 ZIP 압축 파일로 만들기
#zipfile 모듈은 ZIP 파일 관련 다양한 기능들을 제공한다.
from zipfile import*
 
def compressZip(zipname, filename):
    print('[%s] -> [%s] 압축...' %(filename, zipname))
    with ZipFile(zipname, 'w') as zipfile: 
        #ZIP파일 기록을 위하여 ZipFile 쓰기모드로 생성
        zipfile.write(filename) 
        #zip파일에 filename의 데이터를 zip으로 압축한다.
        
    print('압축이 끝났습니다.')
    
filename = './files/stockcode.txt'
zipname = filename + '.zip'
compressZip(zipname, filename)
```

```python
from zipfile import ZipFile

def compressZip(zipname, filename):
    with ZipFile(zipname,'w') as zip:
        zip.write(filename)
        
compressZip('아이와바다.txt.zip','아이와바다.txt')
```

```python
def extractZip(zipname):
    with ZipFile(zipname,'r') as zip:
        zip.extractall()
        
extractZip('아이와바다.txt.zip')
```

```python
#디렉터리를 하나의 ZIP 압축 파일로 만들기 - 예시1
#폴더의 서브디렉터리와 파일들을 탐색하고 결과를 리스트로 저장한다 (files)
#files 리스트에 저장된 file를 모두 zip한다
from zipfile import *
import os

def compressAll(zipname, folder):
    print('[%s] -> [%s] 압축...' %(folder, zipname))
    with ZipFile(zipname, 'w') as zipfile:
        for dirname, subdir, files in os.walk(folder):
            for file in files:
                zipfile.write(os.path.join(dirname,file))

folder = 'test2'
zipname = folder + '.zip'
compressAll(zipname, folder)
```

```python
#디렉터리를 하나의 ZIP 압축 파일로 만들기 - 예시2
from zipfile import ZipFile
import os

def get_all_file_paths(directory):
    #initialize empty file paths list
    file_paths=[]
    
    #crawling through directory and subdirectories
    for root, directories, files in os.walk(directory):
        for filename in  files: 
            #join the two strings in order to form the full file
            filepath = os.path.join(root,filename)
            file_paths.append(filepath)
    #returning all file paths
    return file_paths

def main():
    #path to folder which needs to be zipped
    directory = './test2'
    
    #calling function to get all file paths in the directory
    file_paths = get_all_file_paths(directory)
    
    #printing the list of all files to be zipped
    print('Following files will be zipped:')
    for file_name in file_paths:
        print(file_name)
        
    #writing files to a zipfile
    with ZipFile('test22.zip','w') as zip:
        for file in file_paths:
            zip.write(file)
main()
```
